**EXPLORANDO A FUNÇÃO AGRUPAR**

```python
# Agrupando o banco de dados a partir da raça/etnia dos alunos e gênero
df2 = df.groupby('gender')[['race/ethnicity']].count()
df2
# Aqui abaixo vemos a quantidade de generos por raça e etinia
```

```python
# Agrupando o banco de dados a partir do do Nível de Escolaridade dos pais e se possuem curso preparatório para os testes
df3 = df.groupby('parental level of education')[['test preparation course']].count()
df3
# Abaixo vemos o nivel de escolaridade dos pais em comparação a participação em preparatorios
```

```python
# Agrupando o banco de dados a partir da participação em um curso preparatório para os testes e seu gênero
df4 = df.groupby('test preparation course')[['gender']].count()
df4
# Abaixo vemos o nivel de participação em curso preparatorios em comparação ao gênero da pessoa
```

**EXPLORANDO A PLOTAGEM GRÁFICA:**

**Plotando gráficos com a média matemática**

```python
# Plotando um gráfico com a média matemática para cada grupo disponivel
medmat = df.groupby("gender")["math score"].mean().reset_index()

medmat.plot(kind="bar", x="gender", y="math score")
```

```python
# média matemática para raça e etnia
medmat2 = df.groupby("race/ethnicity")["math score"].mean().reset_index()

medmat2.plot(kind="bar", x="race/ethnicity", y="math score", color = 'red')
```

```python
# média matemática para nivel de escolaridade dos pais
medmat3 = df.groupby("parental level of education")["math score"].mean().reset_index()

medmat3.plot(kind="bar", x="parental level of education", y="math score", color = 'green')
```

**Plotando gráficos com a nota média em Escrita** *texto em itálico*

```python
medwt = df.groupby("lunch")["writing score"].mean().reset_index()

medwt.plot(kind="barh", x="lunch", y="writing score", color = 'pink')
```

```python
medwt2 = df.groupby("test preparation course")["writing score"].mean().reset_index()

medwt2.plot(kind="barh", x="test preparation course", y="writing score", color = 'black')
```

```python
medwt3 = df.groupby("gender")["writing score"].mean().reset_index()

medwt3.plot(kind="barh", x="gender", y="writing score", color = 'yellow')
```

**Plotando gráficos com a nota média em leitura**

```python
medrd = df.groupby("gender")["reading score"].mean().reset_index()

genders = medrd["gender"]
avg_reading_scores = medrd["reading score"]


plt.plot(genders, avg_reading_scores)

plt.xlabel("Gender")
plt.ylabel("Average Reading Score")
plt.title("Average Reading Score by Gender")

plt.show()
```
