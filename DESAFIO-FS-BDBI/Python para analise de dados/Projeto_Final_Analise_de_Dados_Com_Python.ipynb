**Passo 4 - Agrupamento de Dados e Plotagem de Gráficos**

Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);


```python
# Criando variável para armazenar o numero presente emdepartamento
departmento = df['Department'].value_counts()

# Plotando graficamente e modificando
plt.bar(departmento.index, departmento.values, color = 'red')
plt.xlabel('Departamento')
plt.ylabel('Numero de empregados')
plt.title('Empregados por departamento')
```

Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education)

```python
# Agrupando a coluna educação para plotagem
edu = df.groupby("Education").size()

# Criando e modificando a plotagem
plt.plot(edu.index, edu.values)
plt.xlabel("Nivel de escolaridade")
plt.ylabel("Numero de empregados")
plt.title("Empregador por nivel de educação")
```

Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).

```python
# Variável com os valores da coluna Attrition
atrito = df["Attrition"].value_counts()

# Definindo os valores e cores que aparecerão no gráfico
valores = ["Sem atrito", "Com atrito"]
plt.pie(atrito, labels=valores)
plt.title("Empregados com atrito")
```
